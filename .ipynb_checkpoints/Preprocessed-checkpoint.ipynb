```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectKBest,mutual_info_classif
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.feature_selection import chi2
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import PCA
```

```python
training_set = pd.read_csv("project_data/train_dataset.csv")
y_train = training_set['imdb_score_binned']
```

## Handling categorical variables

#### Countvec

```python
# Number of selected best features 
k = 3
```

```python
train_director_name = np.load("project_data/features_countvec/train_countvec_features_director_name.npy")
train_actor_1_name = np.load("project_data/features_countvec/train_countvec_features_actor_1_name.npy")
train_actor_2_name = np.load("project_data/features_countvec/train_countvec_features_actor_2_name.npy")
a3_countvec = CountVectorizer()
train_actor_3_name = a3_countvec.fit_transform(training_set['actor_3_name']).toarray()

dir_scl = StandardScaler()
a1_scl = StandardScaler()
a2_scl = StandardScaler()
a3_scl = StandardScaler()

dir_pca = PCA(n_components=k, random_state=42)
a1_pca =  PCA(n_components=k, random_state=42)
a2_pca = PCA(n_components=k, random_state=42)
a3_pca = PCA(n_components=k, random_state=42)

dir_name = dir_scl.fit_transform(pd.DataFrame(train_director_name))
a1_name = a1_scl.fit_transform(pd.DataFrame(train_actor_1_name))
a2_name = a2_scl.fit_transform(pd.DataFrame(train_actor_2_name))
a3_name = a3_scl.fit_transform(pd.DataFrame(train_actor_3_name))

dir_train_pca = dir_pca.fit_transform(dir_name,y_train)
dir_trans = pd.DataFrame(dir_train_pca, columns=['dir_' + str(i) for i in range(k)])
a1_train_pca = a1_pca.fit_transform(a1_name,y_train)
a1_trans = pd.DataFrame(a1_train_pca, columns=['a1_' + str(i) for i in range(k)])
a2_train_pca = a2_pca.fit_transform(a2_name,y_train)
a2_trans = pd.DataFrame(a2_train_pca, columns=['a2_' + str(i) for i in range(k)])
a3_train_pca = a3_pca.fit_transform(a3_name,y_train)
a3_trans = pd.DataFrame(a3_train_pca, columns=['a3_' + str(i) for i in range(k)])

dir_a_df = pd.concat([dir_trans, a1_trans, a2_trans, a3_trans], axis=1)
dir_a_df
```

#### doc2vec

```python
train_plot_keywords = np.load("project_data/features_doc2vec/train_doc2vec_features_plot_keywords.npy")
train_genre = np.load("project_data/features_doc2vec/train_doc2vec_features_genre.npy")

plot_scl = StandardScaler()
genre_scl = StandardScaler()

plot_pca = PCA(n_components=k, random_state=42)
genre_pca = PCA(n_components=k, random_state=42)

plot = plot_scl.fit_transform(pd.DataFrame(train_plot_keywords))
genre = genre_scl.fit_transform(pd.DataFrame(train_genre))

plot_train_pca = plot_pca.fit_transform(plot, y_train)
genre_train_pca = genre_pca.fit_transform(genre, y_train)

plot_trans = pd.DataFrame(plot_train_pca, columns=['plot_' + str(i) for i in range(k)])
genre_trans = pd.DataFrame(genre_train_pca, columns=['genre_' + str(i) for i in range(k)])

plot_genre_df = pd.concat([plot_trans, genre_trans], axis=1)
plot_genre_df
```

#### Fasttext

```python
train_title_embeddings = np.load("project_data/features_fasttext/train_fasttext_title_embeddings.npy")

title_scl = StandardScaler()

title_pca = PCA(n_components=k, random_state=42)

title = title_scl.fit_transform(pd.DataFrame(train_title_embeddings))

title_train_pca = title_pca.fit_transform(title, y_train)

title_df = pd.DataFrame(title_train_pca, columns=['title_' + str(i) for i in range(k)])
title_df
```

#### Converting 'language', 'country', 'content_rating' to numeric

```python
print(training_set['language'].value_counts()[:5]/training_set.shape[0])
print(training_set['country'].value_counts()[:5]/training_set.shape[0])
print(training_set['content_rating'].value_counts()[:5]/training_set.shape[0])
```

```python
lang_Eng = training_set['language'].apply(lambda x: 1 if x == 'English' else 0)
country_USA = training_set['country'].apply(lambda x: 1 if x == 'USA' else 0)
content_rating = training_set['content_rating'].apply(lambda x: 1 if x == 'R'
                                                                 else 2 if x == 'PG-13'
                                                                 else 3 if x == 'PG'
                                                                 else 0)
lcc_df = pd.concat({'lang_Eng': lang_Eng, 'country_USA': country_USA, 'content_rating': content_rating}, axis=1)
lcc_df = lcc_df.apply(lambda feature: pd.factorize(feature)[0])
lcc_df
```

## Numerical variables

```python
numeric_features = training_set.select_dtypes(exclude=['object'])
numeric_fts = pd.DataFrame(numeric_features)
numeric_fts = numeric_fts.drop(['id'], axis=1)
plt.figure(figsize=(12,10))
sns.heatmap(numeric_fts.corr(), annot=True)
```

```python
numeric_fts_copy = numeric_fts.copy()
numeric_fts_copy['a23_facebook_likes'] = numeric_fts['actor_2_facebook_likes'] + numeric_fts['actor_3_facebook_likes']
numeric_fts_copy['num_ppl'] = numeric_fts['num_critic_for_reviews'] + numeric_fts['num_user_for_reviews'] + numeric_fts['num_voted_users']
numeric_fts = numeric_fts_copy.drop(['cast_total_facebook_likes', 'num_critic_for_reviews', 'num_user_for_reviews',
                               'actor_2_facebook_likes', 'actor_3_facebook_likes', 'num_voted_users'], axis=1)
plt.figure(figsize=(12,10))
sns.heatmap(numeric_fts.corr(), annot=True)
```

## Feature engineering


```python
num_scl = StandardScaler()
imdb_score = numeric_fts['imdb_score_binned']
numeric_fts = numeric_fts.drop(['imdb_score_binned'], axis=1)
numeric_fts = pd.DataFrame(num_scl.fit_transform(numeric_fts), columns=numeric_fts.columns)
numeric_fts['imdb_score_binned'] = imdb_score
```

```python
pr_train = pd.concat([dir_a_df, plot_genre_df, title_df, lcc_df, numeric_fts], axis=1)
pr_train.to_csv('preprocessed_data/preprocessed_train.csv', index=False)
```
